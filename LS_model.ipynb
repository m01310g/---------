## 스케일링

```python
import numpy as np
from sklearn.model_selection import train_test_split

X, y = np.arange(120).reshape((30, 4)), list(range(30))
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.33,random_state=1234)
```

### fit: 어떻게 스케일링할지 먼저 결정

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)         #transform을 진행하기위해 fit을 통해 주어진 데이터를 통해 평균과, 분산을 계산해야함
X_test_scaled = scaler.transform(X_test)  #스케일링

X_test_scaled[:5, :]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(X_train)
X_test_scaled = scaler.transform(X_test)

X_test_scaled[:5, :]
```

```python
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_train_scaled[:5, :]
```

### 정규 방정식: 하나의 닫힌 형태로 값이 얻어짐

### 경사 하강법(중요): 미분값 방향으로 x 변경 -> 값이 빠르게 증가

## 경사 하강법 예시

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10,10,1)
f = x ** 2
plt.plot(x,f)
plt.show()
```

```python
x_new= 10
X = []
Y = []
learng_rate= 0.1
for i in range(100):
    old_value= x_new    # x를 게속 바꿔줌
    x_new= old_value-learng_rate*2* old_value
    X.append(old_value-learng_rate* 2 * old_value)
    Y.append(x_new** 2)
plt.plot(x, f)
plt.scatter(X, Y)
plt.show()
```

- LinearRegression: 정규 방정식을 통해 얻은 결과 제공, regression 객체를 만드는 역할까지 -> fit을 통해 학습 -> predict를 통해 예측

## Linear Regression 문제 해결
- MAE가 클수록 학습이 잘 되지 않은 상태

```python
from sklearn.linear_model import LinearRegression

X, y = load_diabetes(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1234)

reg = LinearRegression()
reg = reg.fit(X_train, y_train)
y_pred_train = reg.predict(X_train)

# Mean Absolute Error
print('학습 데이터셋 MAE: ', np.abs(y_pred_train - y_train).mean())

y_pred = reg.predict(X_test)
print('테스트 데이터셋 MAE: ', np.abs(y_pred - y_test).mean())
```

## SGD Regressor

```python
from sklearn.linear_model import SGDRegressor
```

### SGDRegressor은 parameter를 사용
- penalty: 
- max_iter: 
- learning_rate: 
- eta0: 초기 학습률

```python
reg = SGDRegressor(penalty=None, max_iter=10000, learning_rate='constant', eta0=0.1)
reg = reg.fit(X_train, y_train)
y_pred_train = reg.predict(X_train)
print('학습 데이터셋 MAE: ', np.abs(y_pred_train - y_train).mean())

y_pred = reg.predict(X_test)
print('테스트 데이터셋 MAE: ', np.abs(y_pred - y_test).mean())
```
